```python
def fibonacci(n):
    # Start with the first two Fibonacci numbers
    fib_sequence = [0, 1]

    # Generate the rest of the sequence
    for i in range(2, n):
        next_number = fib_sequence[i-1] + fib_sequence[i-2]
        fib_sequence.append(next_number)

    return fib_sequence
```
